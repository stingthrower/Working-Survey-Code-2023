```python
# for independent attractiveness
# the images in the grid are 1-30 left to right top to bottom.
# using this info I can check each individual one along with each individual T or F to see if it's A or N
import os
from os import listdir
Set_A = []
Set_B = []
files = os.listdir('./RESPONSES')
paths = [os.path.join('./RESPONSES/', basename) for basename in files if "txt" in basename]
for path in paths:
    filename = open(path,'r')
    f_list = filename.readlines()
    if f_list[0].strip() == "Set A":
        Set_B.append(path)
    elif f_list[0].strip() == "Set B":
        Set_A.append(path)
        

# independent attractiveness of Set A put in a dict
ind_res_A = {}
for path in Set_A:
    counter = 1
    a1_dict = {}
    filename = open(path, 'r')
    f_list = filename.readlines()
    for i in range(5,10):
        line = f_list[i].split()
        for tf in line:
            if tf == "True":
                a1_dict[counter] = 'A'
                counter += 1
            else:
                a1_dict[counter] = "N"
                counter += 1
    filename.close()
    ind_res_A[path] = a1_dict
a2_dict = {}
for k , v in ind_res_A.items():
    for key, value in v.items():
        if key not in a2_dict:
            a2_dict[key] = value
        else:
            a2_dict[key] += value
a_dict = {}

# fix this
for h,j in a2_dict.items():
    if j.count("A") > 4:
        a_dict[h] = "A"
    else:
        a_dict[h] = "N"
print('Set A:',a_dict)


# independent attractiveness of Set B put in a dict
ind_res_B = {}
for path in Set_B:
    counter = 1
    b1_dict = {}
    filename = open(path, 'r')
    f_list = filename.readlines()
    for i in range(5,10):
        line = f_list[i].split()
        for tf in line:
            if tf == "True":
                b1_dict[counter] = 'A'
                counter += 1
            else:
                b1_dict[counter] = "N"
                counter += 1
    filename.close()
    ind_res_B[path] = b1_dict
b2_dict = {}
for k , v in ind_res_B.items():
    for key, value in v.items():
        if key not in b2_dict:
            b2_dict[key] = value
        else:
            b2_dict[key] += value
b_dict = {}

# fix this
for h,j in b2_dict.items():
    if j.count("A") > 4:
        b_dict[h] = "A"
    else:
        b_dict[h] = "N"
print('Set B:',b_dict)
```

```python
# dictionary where the key is the path to the response and the value is a dictionary of the mean attractiveness.
# The value dictionary is the number of the image and whether it is attractive, "A", or not, "N".
import os
from os import listdir
files = os.listdir('./RESPONSES')
paths = [os.path.join('./RESPONSES/', basename) for basename in files if "txt" in basename]
responses_A = {}
responses_B = {}
group_mean = []
for path in paths:
    filename = open(path,'r')
    f_list = filename.readlines()
    a = []
    for letter in f_list[3]:
        a.append(letter.strip())
    acount = a.count('A')
    mean = acount/150
    group_mean.append(mean)
    num = f_list[4].split()
    full_dict = {}
    for i in range(150):
        full_dict[num[i]] = full_dict.get(num[i],[]) + [a[i]]
    filename.close()
    g_dict = {}
    for i in range(1,31):
        li = full_dict[str(i)]
        if li.count('A')/5 > mean:
            g_dict[i] = 'A'
        else:
            g_dict[i] = 'N'
    if f_list[0].strip() == "Set A":
        responses_A[path] = g_dict
    elif f_list[0].strip() == "Set B":
        responses_B[path] = g_dict 
print('A',responses_A)
print('B',responses_B)
print(sum(group_mean)/16)
```
